```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# Data processing and Cleaning
```

```python
df =pd.read_csv('Diwali Sales Data.csv', encoding='unicode_escape')
```

```python
df.shape
```

```python
df.head(10)
```

```python
df.info()
```

```python
df.columns
```

```python
df.drop(['Status', 'unnamed1'] ,axis=1, inplace=True)
```

```python
df.info()
```

```python
pd.isnull(df)
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df['Amount']= df['Amount'].astype('int')
```

```python
df['Amount'].dtypes
```

```python
df.columns
```

```python
df.rename(columns= {'Marital_Status': 'Shaadi'})
```

```python
# Exploratory Data Analysis
```

```python
# Gender
```

```python
df.columns
```

```python
ax = sns.countplot(x='Gender', data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen =df.groupby(['Gender'], as_index= False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x= 'Gender', y='Amount', data=sales_gen)
```

```python
# Age
```

```python
df.columns
```

```python
ax= sns.countplot(data=df, x='Age Group', hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state=df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending= False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=sales_state, x='State', y='Orders')
```

```python
#Marital_status
```

```python
ax = sns.countplot(x='Marital_Status', data=df)

for bars in ax.containers:
    ax.bar_label(bars)
    sns.set(rc={'figure.figsize':(6,5)})
```

```python
sales_marital_status=df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending= False).head(10)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data=sales_marital_status, x='Marital_Status', y='Amount', hue='Gender')
```

```python
# Occupation
```

```python
ax = sns.countplot(x='Occupation', data=df)

for bars in ax.containers:
    ax.bar_label(bars)
    sns.set(rc={'figure.figsize':(15,5)})
```

```python
sales_state=df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending= False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=sales_state, x='Occupation', y='Amount')
```

```python
# Product Categroy
```

```python
ax = sns.countplot(x='Product_Category', data=df)

for bars in ax.containers:
    ax.bar_label(bars)
    sns.set(rc={'figure.figsize':(15,5)})
```

```python
sales_state=df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending= False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=sales_state, x='Product_Category', y='Amount')
```

```python
# Top Selling Product
```

```python
sales_state=df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending= False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=sales_state, x='Product_ID', y='Orders')
```
